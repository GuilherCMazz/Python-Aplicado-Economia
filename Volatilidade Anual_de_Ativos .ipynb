#Importando Bibliotecas 

```
import numpy as np
import yfinance as yf 
```

# Definindo os Par√¢metros

```
ativos=['VALE3.SA','BRML3.SA','CIEL3.SA']
```

```
pesos=np.array([0.2,0.3,0.1])
portfolio=yf.download(ativos,start='2020-12-19', end='2022-12-19')['Adj Close']
```

```
portfolio.head()
```

```
portfolio_retornos=portfolio.pct_change()
portfolio_retornos
```

```
portfolio_retornos.dropna(inplace=True)
portfolio_retornos
```

```
# Calculando Matriz de Covari√¢ncia

cov_portfolio=portfolio_retornos.cov()
cov_portfolio
```

# Calculando a Volatilidade


```
vol_diaria=np.sqrt(np.dot(pesos.T,np.dot(cov_portfolio,pesos)))
vol_diaria
```

```
vol_anualizada=vol_diaria*(np.sqrt(252))
vol_anualizada
```

# A volatididade desses 3 ativos com os pesos estabelecidos √© de 19% no periodo de 2020 a 2022.
