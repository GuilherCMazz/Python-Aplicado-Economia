O objetivo do estudo de solvência é identificar a razão de solvabilidade do plano de beneficios do RPPS, considerando hipóteses de simulaçao com base em fluxoa de caixa e cenários macroeconomicos.


```
# Definição dos Parâmetros

x_1=((ativo circulante-passivo circulante)/ativo total))

x_2= reservas e lucros suspensos /ativo total

x_3=((lucro líquido + despesas financeiras + imp. de renda)/ativo total))

x_4= patrimonio líquido / exigivel total

x_5=  Vendas / exigivel total
```

```
# Modelo de Altman

z_1= -1.44+4.03*x_2+2.25*x_3-0.14*x_4+0.42*x_5
z_2= -1.84-0.51*x_1+6.32*x_3+0.71*x_4+0.52*x_5

# onde z_1 ou z_2 são o  total de pontos obtidos
```

O ponto critico para esse modelo é 0, ou seja, para z_1 ou z_2 acima de 0, a empresa é solvente, para valores abaixo de 0 é insolvente.
